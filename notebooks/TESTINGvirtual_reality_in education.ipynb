```python
import numpy as np
from scipy.stats import linregress

# Separate the data by VR usage
grouped = df.groupby('vr_usage')

# Store results
results = {}

# Loop through each group (Yes/No for VR usage)
for name, group in grouped:
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(group['hours_per_week'], group['effectiveness_perception'])
    
    # Store the results
    results[name] = {
        'slope': slope,
        'intercept': intercept,
        'r_value': r_value,
        'p_value': p_value,
        'std_err': std_err
    }

# Display the results
for usage, stats in results.items():
    print(f"VR Usage: {usage}")
    print(f"Slope: {stats['slope']}")
    print(f"Intercept: {stats['intercept']}")
    print(f"R-squared: {stats['r_value']**2}")
    print(f"P-value: {stats['p_value']}")
    print(f"Standard Error: {stats['std_err']}\n")
```
