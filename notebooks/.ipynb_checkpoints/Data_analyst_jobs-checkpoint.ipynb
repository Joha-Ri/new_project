```python
import pandas as pd
```

```python
df = pd.read_csv('eu_entry_level_data_analyst_linkedin_jobs_sample.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Fill missing values with their respective modes

mode_mode = df['mode'].mode()[0]  # Calculate mode for 'mode' column
mode_city = df['city'].mode()[0]  # Calculate mode for 'city' column

# Fill the missing values
df['mode'] = df['mode'].fillna(mode_mode)
df['city'] = df['city'].fillna(mode_city)

# Check again for missing values to confirm they are filled
print(df.isnull().sum())
```

```python
# Remove leading and trailing spaces from all object columns
df[df.select_dtypes(['object']).columns] = df.select_dtypes(['object']).apply(lambda x: x.str.strip())
```

```python
# cleaning the description column
df['description'] = df['description'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
```

```python
df['description'] = df['description'].str.lower()
```

```python
print(df[['description']].head())
```

```python
# Calculate the mode of the city column

mode_city = df['city'].mode()[0]  # Get the most frequent city name

# Fill NaN values in the city
df['city'] = df['city'].fillna(mode_city)

# Verify the changes
print(df['city'].isnull().sum())  # Should print 0, indicating all NaN values are replaced
print(df[['city']].head())  # Display the city column to see if changes are applied
```

```python
df.head()
```

```python
# Convert all object columns to lowercase
df[df.select_dtypes(['object']).columns] = df[df.select_dtypes(['object']).columns].apply(lambda x: x.str.lower())

# Display the updated DataFrame to confirm changes
print(df.head())
```

```python
# Function to remove special characters
def clean_text(text):
    # Check if the text is of type str; if not, return as is
    if isinstance(text, str):
        # Replace special characters with nothing
        return text.replace('\n', ' ').replace('\r', '').replace('"', '').replace("'", "").replace("–", "")\
                   .replace("’", "").replace("…”", "").replace("“", "")  # Add other unwanted chars as needed
    return text

# Apply the clean_text function to relevant columns
columns_to_clean = ['job_title', 'contract', 'city', 'country', 'company', 'sector', 
                    'job_posting_title', 'description', 'linkedin_skills']

# Removing special characters from selected columns
for column in columns_to_clean:
    df[column] = df[column].apply(clean_text)

# Display the cleaned DataFrame to confirm changes
print(df.head())
```

### Extract the skills from description column

```python
import spacy
```

```python
# Load spaCy's English NLP model
nlp = spacy.load('en_core_web_sm')

# Define a list of common data analyst skills
common_skills = ['github', 'sql', 'python', 'r', 'tableau', 'power bi', 'data visualization', 'machine learning', 'excel', 'data analysis', 'data mining', 'statistics', 'communication','etl', 'data cleaning']

def extract_skills(description):
    doc = nlp(description)
    found_skills = []

    # Check each word or phrase in the description against the skills list
    for skill in common_skills:
        if skill.lower() in description.lower():
            found_skills.append(skill)

    return ', '.join(found_skills) if found_skills else None

# Apply skill extraction to the DataFrame
df['extracted_skills'] = df['description'].apply(extract_skills)
```

```python
# Display the updated DataFrame with extracted skills
print(df[['job_title', 'description', 'extracted_skills']].head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Load your DataFrame
# df = pd.read_csv('../data/eu_entry_level_data_analyst_linkedin_jobs_sample.csv')

def scrape_salary(city):
    city_formatted = city.replace(" ", "+").lower()  # Format string for URL
    url = f"https://www.payscale.com/research/US/Job=Data_Analyst/Salary/{city_formatted}"

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }
    
    try:
        response = requests.get(url, headers=headers)  # Add headers to the request
        response.raise_for_status()  # Raise an error for bad responses
        soup = BeautifulSoup(response.text, 'html.parser')  # Parse the HTML
        
        # Look for the salary data element
        salary_elem = soup.find('span', class_='pay-detail-value')
        
        if salary_elem:
            salary = salary_elem.text.strip()
            return salary
        else:
            return None  # Return None if no salary found
        
    except Exception as e:
        print(f"Error occurred while scraping data for {city}: {e}")
        return None

# Apply salary scraping to the DataFrame
df['salary'] = df['city'].apply(scrape_salary)

# Display the DataFrame to see results
print(df[['city', 'salary']].head())
```
